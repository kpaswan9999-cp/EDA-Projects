```python
# Create two datasets
churned = banking_customer[banking_customer["churn"] == 1]
not_churned = banking_customer[banking_customer["churn"] == 0]

# Plot side-by-side pie charts
fig, ax = plt.subplots(1, 2, figsize=(14,7))

# Churned Customers
churned["state"].value_counts().plot(
    kind="pie", autopct="%.2f%%", ax=ax[0], title="Churned Customers by State"
)
ax[0].set_ylabel("")

# Not Churned Customers
not_churned["state"].value_counts().plot(
    kind="pie", autopct="%.2f%%", ax=ax[1], title="Not Churned Customers by State"
)
ax[1].set_ylabel("")

plt.show()
```
